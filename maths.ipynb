```python
# given that 

x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,9])
```

```python
#  the feature matrix X (with a column of ones for the bias)
X = np.vstack((np.ones(len(x)), x)).T


X
```

```python
# computing (X^T  * X) AND (X^T * y)
X_transpose = X.T

XTX = np.dot(X_transpose, X)
XTy = np.dot(X_transpose,y)
```

```python
# compute the coefficients using the normal equation 
coefficients = np.dot(np.linalg.inv(XTX),XTy)
```

```python
# extracting b and w
b = coefficients[0]
w = coefficients[1]
```

```python
print ('Computed coefficients:')
print (f"Intercept (b): {b}")
print (f"SLOPE (w): {w}")
```
