v1 = [1,1], v2 = [1,-1]

````python
import numpy as np

# Given vectors
v1 = np.array([1, 1])
v2 = np.array([1, -1])

# Compute dot products
dot_v1_v1 = np.dot(v1, v1)  # dot product of v1 with itself
dot_v1_v2 = np.dot(v1, v2)  # dot product of v1 and v2

# Compute u1 and u2 using Gram-Schmidt orthogonalization
u1 = v1
u2 = v2 - (dot_v1_v2 / dot_v1_v1) * v1

print("Orthogonal vectors:")
print("u1:", u1)
print("u2:", u2)
````

### **Moore-Penrose Pseudoinverse Explanation**
The **Moore-Penrose Pseudoinverse** (denoted as \( B^+ \)) is a generalization of the matrix inverse that works for **non-square** or **singular** matrices. Unlike the regular inverse, which only exists for square matrices with full rank, the pseudoinverse provides a best-fit solution in **least squares problems**.

For a given matrix \( B \), the pseudoinverse satisfies these conditions:
1. \( BB^+B = B \)
2. \( B^+BB^+ = B^+ \)
3. \( (BB^+)^T = BB^+ \)  (Symmetric property)
4. \( (B^+B)^T = B^+B \)  (Symmetric property)

---



---

### **Use of the Moore-Penrose Pseudoinverse**
The pseudoinverse is mainly used in:
1. **Solving linear systems where \( B \) is not square**:
   - Instead of solving \( Bx = y \) with an inverse, we use:
     \[
     x = B^+ y
     \]
     which gives the least-squares solution when \( B \) has no exact solution.

2. **Machine Learning (Least Squares Regression)**:
   - The normal equation for linear regression:
     \[
     \theta = (X^T X)^{-1} X^T y
     \]
     is computed as:
     \[
     \theta = X^+ y
     \]
     when \( X \) is not invertible.

3. **Dimensionality Reduction & PCA**:
   - Used in **Singular Value Decomposition (SVD)**, where \( B^+ \) helps reconstruct data.

4. **Control Systems & Signal Processing**:
   - Used in **optimal filtering** and solving **overdetermined/underdetermined systems**.

---

### **Comparing Pseudoinverse vs. Regular Inverse**
- **Regular inverse** \( B^{-1} \) only exists for **square, full-rank** matrices.
- **Pseudoinverse** \( B^+ \) works for **any** matrix, providing:
  - The exact solution when possible.
  - The best **least-squares** approximation when no exact solution exists.

---

### **Conclusion**
- ✅ Your result is **correct**.
- 🔥 The **Moore-Penrose pseudoinverse** is a powerful tool for solving linear systems, machine learning, and signal processing.
- 💡 If you need to solve **underdetermined** or **overdetermined** systems, use `np.linalg.pinv()` instead of `np.linalg.inv()`.

### **Do We Get the Same Answer Using \( B^{-1} \) vs. \( B^+ \)?**

It depends on whether **\( B \) is square and invertible**. Let's break it down:

#### **Case 1: If \( B \) is a square and full-rank matrix (\( B^{-1} \) exists)**
If \( B \) is an \( n \times n \) invertible matrix, the **pseudoinverse** and the **regular inverse** give the same result:
\[
B^{-1} = B^+
\]
So, solving \( Bx = y \) gives:
\[
x = B^{-1}y = B^+y
\]
✅ **Same answer in this case.**

---

#### **Case 2: If \( B \) is not square or singular (not invertible)**
- If \( B \) is **rectangular** (e.g., \( m \times n \) where \( m \neq n \)), then \( B^{-1} \) **does not exist**.
- If \( B \) is **singular** (determinant is 0), then \( B^{-1} \) **does not exist**.

Here, **only the pseudoinverse \( B^+ \) works**, and it finds a **least-squares solution** instead of an exact solution.

Instead of solving \( Bx = y \) exactly (which may not be possible), we get:
\[
x = B^+ y
\]
✅ **This gives the best possible solution (least-squares approximation).**

---

### **Example in Python**
Let’s test both cases:

#### **Case 1: Square, Invertible Matrix**
```python
import numpy as np

B = np.array([[2, 3], [1, 4]])  # Square, full-rank
y = np.array([5, 7])

# Regular inverse solution
x_inv = np.linalg.inv(B) @ y

# Pseudoinverse solution
x_pinv = np.linalg.pinv(B) @ y

print("Inverse Solution:", x_inv)
print("Pseudoinverse Solution:", x_pinv)
```
✅ Both will return **the same answer**.

---

#### **Case 2: Rectangular Matrix**
```python
B = np.array([[1, 2], [3, 4], [5, 6]])  # 3x2 matrix (not square)
y = np.array([5, 7, 9])

# Try to compute inverse (this will fail)
# x_inv = np.linalg.inv(B) @ y  # ERROR: Matrix is not square

# Compute pseudoinverse solution
x_pinv = np.linalg.pinv(B) @ y

print("Pseudoinverse Solution:", x_pinv)
```
❌ `np.linalg.inv(B)` **fails** because \( B \) is not square.  
✅ `np.linalg.pinv(B) @ y` works and gives a **least-squares solution**.

---

### **Key Takeaways**
- **If \( B \) is invertible** → \( B^+ = B^{-1} \), so **both methods give the same answer**.
- **If \( B \) is singular or rectangular** → \( B^{-1} \) **does not exist**, but \( B^+ \) provides the **best approximation**.

### **Conclusion**
For solving \( Bx = y \):
- If \( B \) is **square and invertible**, use **\( B^{-1} \)** or \( B^+ \).
- If \( B \) is **not invertible or rectangular**, **only \( B^+ \) works**.

🔥 **In machine learning and least squares problems, \( B^+ \) is preferred because it's more general and always works!**

Movie recommendation using svd 

````python
import numpy as np

# Step 1: Matrix Factorization (SVD)
x = np.array([[3, 4, 0, 1],
              [4, 1, 0, 2],
              [0, 1, 3, 2],
              [1, 0, 0, 2]])

# Perform SVD
U, S, Vt = np.linalg.svd(x, full_matrices=False)

# Convert S from a vector to a diagonal matrix
S_matrix = np.diag(S)

# Reconstruct the matrix
reconstructed_matrix = np.dot(U, np.dot(S_matrix, Vt))

# Step 2: Fill missing ratings (zeros in original matrix)
filled_matrix = np.where(x == 0, reconstructed_matrix, x)

# Step 3: Recommend movies (highest predicted ratings for each user)
recommended_movies = np.argmax(filled_matrix, axis=1)

# Print the reconstructed matrix and the recommended movies
print("Reconstructed Matrix:\n", reconstructed_matrix)
print("\nFilled Matrix (with predicted values):\n", filled_matrix)

print("\nRecommended Movies for each User (0-based index):")
for i, movie in enumerate(recommended_movies):
    print(f"User {i+1} should watch movie {movie+1} (Predicted rating: {filled_matrix[i, movie]:.2f})")
````

 Music Recommendation System
Problem Statement:
You want to create a music recommendation system where you recommend songs to users based on their past listening preferences. You have a user-song rating matrix where each row represents a user, each column represents a song, and the values are ratings (1-5) the user has given to a song. Like the movie example, some users haven’t rated all the songs, so your goal is to predict these missing ratings.

````python
rating = np.array([[4,5,0,2,0],[5,0,3,0,4],[3,4,0,5,4],[0,2,4,3,0]])

U, S , Vt = np.linalg.svd(rating,full_matrices=False)
````

````python
U, S, Vt = np.linalg.svd(rating, full_matrices=False)

# Display the decomposed matrices
print("U matrix:\n", U)
print("Singular values (S):\n", S)
print("V transpose (Vt):\n", Vt)
````

````python
# Keep only the top 2 singular values
k = 2
S_k = np.diag(S[:k])

# Reconstruct the matrix with the top k singular values
reconstructed_matrix = np.dot(np.dot(U[:, :k], S_k), Vt[:k, :])

print("Reconstructed Matrix:\n", reconstructed_matrix)
````

````python
# Recommend top 3 songs for each user
recommendations = np.argsort(reconstructed_matrix, axis=1)[:, -3:]

print("Recommended songs for each user:")
for user_id, recommended_songs in enumerate(recommendations):
    print(f"User {user_id + 1}: Songs {recommended_songs + 1}")
````
